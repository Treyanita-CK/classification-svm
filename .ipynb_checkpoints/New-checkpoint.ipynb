```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Membaca data dari file Excel
data = "Kuisioner.xlsx"
df = pd.read_excel(data, sheet_name="Sheet2")

# Memisahkan fitur dan label
X = df.drop(columns=['label']).values
y = df['label'].values

# Membagi data menjadi data train dan test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Menstandarisasi data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Menyiapkan grid search untuk mencari parameter terbaik
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],
    'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
    'kernel': ['rbf']  # Hanya kernel RBF
}

svm = SVC()

# Mencari parameter terbaik dengan GridSearchCV
grid = GridSearchCV(svm, param_grid, refit=True, verbose=3, cv=5)
grid.fit(X_train, y_train)

# Menampilkan parameter terbaik
print("Best parameters found:")
print(grid.best_params_)
```

```python
## lambda ##
file_path = 'hasil_kernel_rbf_3.xlsx'

data = pd.read_excel(file_path, sheet_name='Sheet1')

X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

y[y == 0] = -1

lambdas = [0.1, 0.5, 10, 50, 100]

C_values = [1.0 / lamb for lamb in lambdas]

kf = KFold(n_splits=5, shuffle=True, random_state=42)

param_grid = {'C': C_values, 'kernel': ['rbf']}
grid_search = GridSearchCV(SVC(), param_grid, cv=kf, scoring='accuracy', return_train_score=False)

grid_search.fit(X, y)

cv_results = grid_search.cv_results_

results = {
    'lambda': [],
    'fold': [],
    'accuracy': []
}

for i, (C, mean_score) in enumerate(zip(cv_results['param_C'], cv_results['mean_test_score'])):
    lambda_value = 1.0 / C
    
    results['lambda'].extend([lambda_value] * 5)
    results['fold'].extend(range(1, 6))
    results['accuracy'].extend([mean_score] * 5) 

results_df = pd.DataFrame(results)

average_accuracies = results_df.groupby('lambda')['accuracy'].mean().reset_index()
average_accuracies.columns = ['lambda', 'average_accuracy']

average_accuracies['average_accuracy'] *= 100

highest_percentage = average_accuracies.loc[average_accuracies['average_accuracy'].idxmax()]

print("Hasil Akurasi per Fold dan Lambda:")
print(results_df)

print("\nRata-rata Akurasi per Lambda (Persentase):")
print(average_accuracies)

print("\nNilai Persentase Tertinggi:")
print(highest_percentage)
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Plot grafik line
plt.figure(figsize=(10, 6))
plt.plot(average_accuracies['lambda'], average_accuracies['average_accuracy'], marker='o', linestyle='-', color='b')

# Atur label sumbu x dan y
plt.xlabel('Lambda')
plt.ylabel('Rata-rata Akurasi (%)')
plt.title('Rata-rata Akurasi per Lambda')

# Atur sumbu x untuk menampilkan nilai lambda tanpa format pangkat
plt.xscale('log')
plt.gca().xaxis.set_major_formatter(ticker.ScalarFormatter())

# Tampilkan grid
plt.grid(True)

# Tampilkan grafik
plt.show()
```

```python
### kernel RBF ###

import pandas as pd
import numpy as np

data = "Kuisioner.xlsx"
df = pd.read_excel(data, sheet_name="Sheet2")

X = df.drop(columns=['label']).values
y = df['label'].values

def rbf_kernel(x_i, x_j, gamma):
    """Menghitung nilai kernel RBF antara dua vektor fitur"""
    distance_squared = np.sum((x_i - x_j) ** 2)
    return np.exp(-gamma * distance_squared)

def compute_kernel_matrix(X, gamma):
    """Menghitung matriks kernel RBF untuk seluruh data"""
    n_samples = X.shape[0]
    kernel_matrix = np.zeros((n_samples, n_samples))  # Perbaiki kesalahan ketik di sini
    
    for i in range(n_samples):
        for j in range(n_samples):
            kernel_matrix[i, j] = rbf_kernel(X[i], X[j], gamma)
    
    return kernel_matrix

gamma = 0.1 

kernel_matrix = compute_kernel_matrix(X, gamma)

kernel_df = pd.DataFrame(kernel_matrix, columns=[f'Sample_{i}' for i in range(kernel_matrix.shape[1])])

# Simpan DataFrame ke file Excel baru
output_file = "Matriks_Kernel_RBF.xlsx"
kernel_df.to_excel(output_file, sheet_name='Kernel Matrix', index=False)

print(f'Matriks Kernel RBF telah disimpan di {output_file}')
```

```python
import pandas as pd
import numpy as np

# Muat data dari file Excel
data = "Kuisioner.xlsx"
df = pd.read_excel(data, sheet_name="Sheet2")

# Pisahkan fitur dan label
X = df.drop(columns=['label']).values
y = df['label'].values

# Mengubah semua nilai 0 menjadi 1
y = np.where(y == 0, 1, y)

def rbf_kernel(x_i, x_j, gamma):
    """Menghitung nilai kernel RBF antara dua vektor fitur"""
    distance_squared = np.sum((x_i - x_j) ** 2)
    return np.exp(-gamma * distance_squared)

def compute_kernel_matrix(X, gamma):
    """Menghitung matriks kernel RBF untuk seluruh data"""
    n_samples = X.shape[0]
    kernel_matrix = np.zeros((n_samples, n_samples))
    
    for i in range(n_samples):
        for j in range(n_samples):
            kernel_matrix[i, j] = rbf_kernel(X[i], X[j], gamma)
    
    return kernel_matrix

def compute_hessian_matrix(X, y, gamma, lambda_):
    """Menghitung matriks Hessian untuk SVM"""
    n_samples = X.shape[0]
    kernel_matrix = compute_kernel_matrix(X, gamma)
    hessian_matrix = np.zeros((n_samples, n_samples))
    
    for i in range(n_samples):
        for j in range(n_samples):
            hessian_matrix[i, j] = y[i] * y[j] * (kernel_matrix[i, j] + lambda_**2)
    
    return hessian_matrix

gamma = 0.1  
lambda_ = 0.5  

# Hitung matriks Hessian
hessian_matrix = compute_hessian_matrix(X, y, gamma, lambda_)

print(hessian_matrix)
```

```python
### alpha, delta-alpha, learning-rate, epsilon ###

import pandas as pd
import numpy as np

data = "Kuisioner.xlsx"
df = pd.read_excel(data, sheet_name="Sheet2")

X = df.drop(columns=['label']).values
y = df['label'].values

y = np.where(y == 0, 1, y)

def rbf_kernel(x_i, x_j, gamma):
    """Menghitung nilai kernel RBF antara dua vektor fitur"""
    distance_squared = np.sum((x_i - x_j) ** 2)
    return np.exp(-gamma * distance_squared)

def compute_kernel_matrix(X, gamma):
    """Menghitung matriks kernel RBF untuk seluruh data"""
    n_samples = X.shape[0]
    kernel_matrix = np.zeros((n_samples, n_samples))
    
    for i in range(n_samples):
        for j in range(n_samples):
            kernel_matrix[i, j] = rbf_kernel(X[i], X[j], gamma)
    
    return kernel_matrix

def compute_hessian_matrix(X, y, gamma, lambda_):
    """Menghitung matriks Hessian untuk SVM"""
    n_samples = X.shape[0]
    kernel_matrix = compute_kernel_matrix(X, gamma)
    hessian_matrix = np.zeros((n_samples, n_samples))
    
    for i in range(n_samples):
        for j in range(n_samples):
            hessian_matrix[i, j] = y[i] * y[j] * (kernel_matrix[i, j] + lambda_**2)
    
    return hessian_matrix


gamma = 0.1  
lambda_ = 0.5  

hessian_matrix = compute_hessian_matrix(X, y, gamma, lambda_)

learning_rate = 1 / np.max(hessian_matrix)
print(f"Learning rate (1 / nilai terbesar dari matriks Hessian): {learning_rate}")

alpha = np.zeros_like(y)

epsilon = learning_rate * np.dot(hessian_matrix, alpha)

C = 1
alpha_i = 0 # asumsi awal
epsilon_i = 0

print("Nilai Epsilon:")
print(epsilon)

delta_alpha_i = min(max(learning_rate * (1 - epsilon_i), - alpha_i), C - alpha_i)
alpha_j = alpha_i + delta_alpha_i
print(f"delta alpha : {delta_alpha_i}")
print(f"alpha : {alpha_j}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix

# Load the Hessian Kernel Matrix
data = "Data Lama.xlsx"
df_hessian = pd.read_excel(data, sheet_name="Sheet3")

# Check columns
print("Columns in the Excel sheet:", df_hessian.columns)

# Assume the last column is the label
X = df_hessian.drop(columns=['label']).values
y = df_hessian['label'].values

# Modify labels if necessary
y = np.where(y == 0, 1, y)

# Parameters for prediction
alpha = 0.9 # Given alpha
b = 0.03  # Given bias

# Define the custom decision function
def decision_function(X, alpha, y, b):
    return np.dot(X * alpha * y[:, np.newaxis], np.ones(X.shape[1])) + b

# Predict using the decision function
predictions = decision_function(X, alpha, y, b)
predicted_labels = np.sign(predictions)

# Check lengths of y and predicted_labels
print("Length of actual labels (y):", len(y))
print("Length of predicted labels:", len(predicted_labels))

# Ensure lengths match before creating the DataFrame
if len(y) != len(predicted_labels):
    raise ValueError("Length of actual labels and predicted labels do not match.")

# Compare with actual labels
comparison = pd.DataFrame({
    'Actual': y,
    'Predicted': predicted_labels
})

# Count the number of correct predictions
correct_predictions = np.sum(comparison['Actual'] == comparison['Predicted'])
total_predictions = len(comparison)
accuracy = correct_predictions / total_predictions

print("Predictions:\n", comparison)
print(f"Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
conf_matrix = confusion_matrix(y, predicted_labels)
print("Confusion Matrix:\n", conf_matrix)

# Save the comparison to Excel
output_file = "Prediksi_SVM_Hessian.xlsx"
comparison.to_excel(output_file, index=False)
#print(f'Comparison saved to {output_file}')
```
