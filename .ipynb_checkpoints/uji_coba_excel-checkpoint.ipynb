```python
# matriks hessian
import numpy as np

# Definisikan matriks yang diberikan
matrix = np.array([
    [0.974, 0.997, 0.990, 0.977, 0.983, 0.987, 0.994],
    [0.991, 0.978, 0.957, 0.979, 0.990, 0.983, 0.974],
    [0.989, 0.982, 0.955, 0.982, 0.993, 0.987, 0.974],
    [0.995, 0.986, 0.974, 0.982, 0.989, 0.986, 0.990],
    [0.976, 0.982, 0.979, 0.993, 0.984, 0.972, 0.983]
])

# Fungsi untuk menghitung turunan kedua secara numerik
def numerical_hessian(matrix):
    n = matrix.shape[0]
    hessian = np.zeros((n, n))
    for i in range(1, n-1):
        for j in range(1, n-1):
            hessian_xx = (matrix[i+1, j] - 2*matrix[i, j] + matrix[i-1, j])
            hessian_yy = (matrix[i, j+1] - 2*matrix[i, j] + matrix[i, j-1])
            hessian_xy = (matrix[i+1, j+1] - matrix[i+1, j-1] - matrix[i-1, j+1] + matrix[i-1, j-1]) / 4.0
            hessian[i, j] = hessian_xx + hessian_yy + hessian_xy
    return hessian

# Hitung matriks Hessian numerik
hessian_matrix = numerical_hessian(matrix)

# Tampilkan matriks Hessian
print("Matriks Hessian:")
print(hessian_matrix)

# Simpan ke file Excel
excel_filename = 'hasil_kernel_rbf.xlsx'
sheet_name = 'Sheet2'
df = pd.DataFrame(hessian_matrix)
df.to_excel(excel_filename, sheet_name=sheet_name, index=False, header=False)

print(f"Hasil matriks Hessian telah disimpan ke dalam file '{excel_filename}', sheet '{sheet_name}'.")
```

```python
import numpy as np
import pandas as pd

# Matriks kernel K
K = np.array([
    [0.974, 0.997, 0.990, 0.977, 0.983, 0.987, 0.994],
    [0.991, 0.978, 0.957, 0.979, 0.990, 0.983, 0.974],
    [0.989, 0.982, 0.955, 0.982, 0.993, 0.987, 0.974],
    [0.995, 0.986, 0.974, 0.982, 0.989, 0.986, 0.990],
    [0.976, 0.982, 0.979, 0.993, 0.984, 0.972, 0.983],
    [0.984, 0.968, 0.965, 0.972, 0.983, 0.963, 0.974],
    [0.957, 0.965, 0.982, 0.974, 0.964, 0.946, 0.977]
])

# Vektor kelas y
y = np.array([-1, -1, -1, 1, 1, -1, 1])

# Parameter lambda
lambda_val = 0.5
lambda_sq = lambda_val ** 2

# Hitung produk luar dari vektor kelas y
yyT = np.outer(y, y)

# Kalikan setiap elemen dari kernel dengan lambda_sq
K_lambda = K * lambda_sq

# Hitung matriks Hessian H
H = yyT * K_lambda

# Cetak hasil matriks Hessian
print("Matriks Hessian:")
print(H)

# Simpan hasil ke file Excel
df = pd.DataFrame(H)
df.to_excel('matrix_hessian_testing.xlsx', index=False, header=False)
```

```python
import numpy as np
from sklearn import svm
from sklearn.metrics import classification_report, accuracy_score

# Matriks kernel K
K = np.array([
    [1.000, 0.986, 0.965, 0.987, 0.996, 0.991, 0.985],
    [0.986, 1.000, 0.987, 0.984, 0.991, 0.995, 0.996],
    [0.965, 0.987, 1.000, 0.967, 0.974, 0.968, 0.994],
    [0.987, 0.984, 0.967, 1.000, 0.989, 0.984, 0.982],
    [0.996, 0.991, 0.974, 0.989, 1.000, 0.991, 0.989],
    [0.991, 0.995, 0.968, 0.984, 0.991, 1.000, 0.986],
    [0.985, 0.996, 0.994, 0.982, 0.989, 0.986, 1.000]
])

# Vektor kelas y
y = np.array([-1, -1, 1, -1, 1, -1, -1])

X_train = K
y_train = y

# Parameter SVM
C = 10
gamma = 0.01

# Buat model SVM dengan kernel RBF
model = svm.SVC(C=C, kernel='rbf', gamma=gamma)

# Latih model
model.fit(X_train, y_train)

# Prediksi menggunakan model yang telah dilatih
y_pred = model.predict(X_train)

# Evaluasi model
accuracy = accuracy_score(y_train, y_pred)
report = classification_report(y_train, y_pred)

print("Akurasi:", accuracy)
print("Laporan Klasifikasi:\n", report)
```

```python
### RBF Kernel training testing ###

import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

data = 'hasil_kernel_rbf_2.xlsx'
df = pd.read_excel(data, sheet_name='Sheet1')

X = df.drop(columns=['label']).values
y = df['label'].values

# split menjadi data train and test
# Split data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# inisialiasasi model SVM dengan
best_param = {'C': 1, 'gamma': 0.01}
svm_rbf = SVC(kernel='rbf', C=best_param['C'], gamma=best_param['gamma'])

#training
svm_rbf.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = svm_rbf.predict(X_train)

# Hitung confusion matrix
#cm_train = confusion_matrix(y_train, y_pred_train)
#print('Confusion matrix untuk data training:')
#print(cm_train)

# Tampilkan juga classification report jika diperlukan
#print('Classification report untuk data training:')
#print(classification_report(y_train, y_pred_train))

#prediksi data uji
y_pred = svm_rbf.predict(X_test)

# evaluasi
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# confussion matrix
print('Confusion matrix:')
print(confusion_matrix(y_test, y_pred))

#classification
print('Classification report:')
print(classification_report(y_test, y_pred))

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
classes = np.unique(y)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)

fmt = 'd'
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], fmt),
             ha="center", va="center",
             color="white" if cm[i, j] > thresh else "black")

plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

```

```python
### RBF Kernel training testing ###

import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd

data = 'Kuisioner.xlsx'
df = pd.read_excel(data, sheet_name='kuisioner')

X = df.drop(columns=['label']).values
y = df['label'].values

# split menjadi data train and test
# Split data menjadi train dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# inisialiasasi model SVM dengan
best_param = {'C': 1, 'gamma': 0.01}
svm_rbf = SVC(kernel='rbf', C=best_param['C'], gamma=best_param['gamma'])

#training
svm_rbf.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = svm_rbf.predict(X_train)

# Hitung confusion matrix
#cm_train = confusion_matrix(y_train, y_pred_train)
#print('Confusion matrix untuk data training:')
#print(cm_train)

# Tampilkan juga classification report jika diperlukan
#print('Classification report untuk data training:')
#print(classification_report(y_train, y_pred_train))

#prediksi data uji
y_pred = svm_rbf.predict(X_test)

# evaluasi
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# confussion matrix
print('Confusion matrix:')
print(confusion_matrix(y_test, y_pred))

#classification
print('Classification report:')
print(classification_report(y_test, y_pred))

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
classes = np.unique(y)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)

fmt = 'd'
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], fmt),
             ha="center", va="center",
             color="white" if cm[i, j] > thresh else "black")

plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
confusion_matrix_path = 'confusion_matrix.png'
plt.savefig(confusion_matrix_path)
plt.show()

# Create pie chart for classification results
labels = np.unique(y)
sizes = [np.sum(y_pred == label) for label in labels]

# Convert counts to percentages
total_count = len(y_pred)
percentages = [size / total_count * 100 for size in sizes]

plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=[f'{label} ({percent:.1f}%)' for label, percent in zip(labels, percentages)],
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(np.arange(len(labels))))
plt.title('Classification Result Percentages')
pie_chart_path = 'classification_pie_chart.png'
plt.savefig(pie_chart_path)  # Save pie chart as PNG
plt.show()

# Save model and scaler
model_filename = 'svm_rbf_model_2.pkl'
scaler_filename = 'scaler.pkl'

joblib.dump(svm_rbf, model_filename)
joblib.dump(scaler, scaler_filename)

#print(f'Model and scaler saved to {model_filename} and {scaler_filename}')
```
